```python
import matplotlib.pyplot as plt
import numpy as np

# Weight Values: [0, 1, 1]
x1 = np.linspace(-2, 2, 100)
y1 = -x1  # Decision boundary equation: x + y = 0

# Weight Values: [0, 1, -1]
x2 = np.linspace(-2, 2, 100)
y2 = x2  # Decision boundary equation: x - y = 0

# Weight Values: [-1, 1, 1]
x3 = np.linspace(-2, 2, 100)
y3 = 1 - x3  # Decision boundary equation: x + y = 1

plt.figure(figsize=(8, 6))
plt.plot(x1, y1, label="[0, 1, 1] Decision Boundary", linestyle='--')
plt.plot(x2, y2, label="[0, 1, -1] Decision Boundary", linestyle='-.')
plt.plot(x3, y3, label="[-1, 1, 1] Decision Boundary", linestyle=':')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Feature X')
plt.ylabel('Feature Y')
plt.title('Decision Boundaries')
plt.legend()
plt.grid(True)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```
